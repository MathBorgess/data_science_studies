```python
import pandas as pd

spaceship_dataset = pd.read_csv('./data/train.csv')

# As a passenger, would you be transported to another dimension?
spaceship_dataset.head()
```

```python
spaceship_dataset[['CabinDeck','CabinNumber','CabinSide' ]] = spaceship_dataset['Cabin'].str.split('/', expand=True)
spaceship_dataset.head()
```

```python
spaceship_dataset[['GroupId','InGroupId']] = spaceship_dataset['PassengerId'].str.split('_', expand=True)
spaceship_dataset.head()
```

```python
spaceship_dataset['GroupLength'] = spaceship_dataset['GroupId'].map(spaceship_dataset.groupby(['GroupId'])['Name'].count())
spaceship_dataset.head()
```

```python
spaceship_dataset[['FirstName', 'Surname']] = spaceship_dataset['Name'].str.split(' ', expand=True)
spaceship_dataset.head()
```

```python
spaceship_dataset['FamiliarGroup'] = spaceship_dataset[spaceship_dataset['GroupLength'] > 1]['GroupId'].map(spaceship_dataset[spaceship_dataset['GroupLength'] > 1].groupby(['GroupId'])['Surname'].unique())
spaceship_dataset['FamiliarGroup']
```

```python
spaceship_dataset['FamiliarGroupBinary'] = spaceship_dataset['FamiliarGroup'].fillna('').apply(lambda x: 1 if len(x) == 1 else 0)
spaceship_dataset['FamiliarGroupBinary']
```

```python
training_set = spaceship_dataset.drop(columns=['Name', 'FirstName', 'Surname', 'Cabin', 'FamiliarGroup'])
training_set.isna().sum()
```

```python
training_dummie_set = pd.get_dummies(training_set,columns=['HomePlanet', 'Destination', 'CabinDeck','CabinSide'],  dtype=int, drop_first=True)
training_dummie_set[['CryoSleep','VIP','Transported']] = training_dummie_set[['CryoSleep','VIP','Transported']].replace({True: 1, False: 0})
training_dummie_set.head()
```

```python
training_dummie_set.describe()
```

```python
from sklearn.impute import KNNImputer
impute = KNNImputer(n_neighbors=round(len(training_dummie_set.columns)**1/2))
training_filled_set = impute.fit_transform(training_dummie_set)
training_filled_set = pd.DataFrame(training_filled_set, columns=training_dummie_set.columns).drop(columns=['PassengerId', 'GroupId', 'InGroupId'])
training_filled_set.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

colormap = plt.cm.viridis
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(training_filled_set.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
len(training_filled_set.columns)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

params = {
    'n_estimators' : [100,200,500],
    'max_depth' : range(1,len(training_filled_set.columns),3),
    'max_features': ['log2','sqrt']
}

random_search = GridSearchCV(RandomForestClassifier(),params,cv=10)

random_search.fit(training_filled_set.drop(columns=['Transported']), training_filled_set['Transported'])
```

```python

df = pd.DataFrame(random_search.cv_results_)[["param_max_features", "param_max_depth",'param_n_estimators',"mean_test_score"]]
df
```

```python
sns.catplot(
    data=df,
    kind="bar",
    x="param_max_depth",
    y="mean_test_score",
    hue="param_max_features"
)
plt.show()
```

```python

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

model = RandomForestClassifier(n_estimators= 500, max_depth= random_search.best_params_['max_depth'], max_features=random_search.best_params_['max_features'])
X_train, X_test, y_train, y_test = train_test_split(training_filled_set.drop(columns=['Transported']), training_filled_set['Transported'], test_size = 0.3)
model.fit(X= X_train, y= y_train)
```

```python
accuracy_score(y_test, model.predict(X_test))
```
